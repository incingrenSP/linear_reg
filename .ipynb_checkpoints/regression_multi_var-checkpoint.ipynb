```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/data2.txt')
df.head()
```

```python
df['Size_normal'] = (df['Size'] - df['Size'].mean()) / df['Size'].std()
df['Bedrooms_normal'] = (df['Bedrooms'] - df['Bedrooms'].mean()) / df['Bedrooms'].std()
df['Price_normal'] = (df['Price'] - df['Price'].mean()) / df['Price'].std()
```

```python
df.head()
```

```python
plt.figure()
graph = plt.axes(projection='3d')
graph.scatter3D(df['Size'], df['Bedrooms'], df['Price'])
```

```python
plt.figure()
graph = plt.axes(projection='3d')
graph.scatter3D(df['Size_normal'], df['Bedrooms_normal'], df['Price_normal'])
```

```python
X = np.array([df['Size_normal'], df['Bedrooms_normal']])
y = df['Price_normal']
```

```python
def predict(w, b, X):
    # print(f'w:{w.shape}, X:{X.shape}')
    res = np.dot(w, X) + b
    return res
```

```python
def total(yp, y, size):
    comp = (yp - y)**2
    return comp.sum()
```

```python
def gradient(w, b, yp, y, size):
    w_g, b_g = np.zeros(2), np.float64(0.0)
    w_fin, b_fin = np.zeros(2), np.float64(0.0)
    
    for i in range(w.size):
        w_g[i] = np.dot((yp - y), X[i]).sum()
    b_g = (yp - y).sum()

    for j in range(w.size):
        w_fin[j] = (1/size) * w_g[j]
    b_fin = (1/size) * b_g
    
    return w_fin, b_fin
```

```python
def linear_regression(X, y):
    w = np.array([1,1])
    b = np.float64(1.0)
    J_all = np.array([])

    for iteration in range(max_iterations):
        f_pred = predict(w, b, X)
        J = (1/2*m) * total(f_pred, y, m)
        J_all = np.append(J_all, J)
    
        w_gradient, b_gradient = gradient(w, b, f_pred, y, m)
        w = w - a * w_gradient
        b = b - a * b_gradient
    return w, b, J_all
```

```python
m = X[0].size
a = 0.01
max_iterations = 300

w, b, J = linear_regression(X, y)
w, b
```

```python
y_predict = w[0] * df['Size_normal'] + w[1] * df['Bedrooms_normal'] + b
df['Prediction_normal'] = y_predict

y_predict = y_predict * df['Price'].std() + df['Price'].mean()
df['Prediction'] = y_predict.astype(np.int64)
```

```python
df.head()
```

```python
df['Error'] = (df['Price'] - df['Prediction']) / df['Price'] * 100
```
